```python
import nextnanopy as nn
import os
```

```python
# example 2 plotting data files
output_path = r'C:\Users\splatt\Documents\nextnano\Output\example'
bias_folder = r'bias_0000'
bandedges_file_name = 'bandedges.dat'
# Combine paths (using os.path.join - recommended)
full_path = os.path.join(output_path, bias_folder, bandedges_file_name)
bandedges_path = r'C:\Users\splatt\Documents\nextnano\Output\example\bias_00000\bandedges.dat'
print(f"Full path: {full_path}")
datafile_1d = nn.DataFile(bandedges_path,product='nextnano++')
```

```python
print(datafile_1d)
datafile_1d.plot(y_axis_name='Energy')
```

```python
datafile_1d.coords
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1)
x = datafile_1d.coords['x']
gamma = datafile_1d.variables['Gamma']
efermi = datafile_1d.variables['electron_Fermi_level']
ax.plot(x.value, gamma.value, label='Gamma Bandedge')
ax.plot(x.value,efermi.value, label='Electron Fermi Level')
ax.set_xlabel(x.label)
ax.set_ylabel(gamma.label)
ax.legend()
fig.tight_layout()
```
